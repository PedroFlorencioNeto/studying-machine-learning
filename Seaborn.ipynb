<b> How to import Seaborn? </b>

```python
import seaborn as sns
%matplotlib inline
```

<b> How to plot a histogram? </b>

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.distplot(tips['tip'],kde=False,bins=15)
```

<b> How to plot a histogram and correlaction of 2 predictors? </b>

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

<b> How to perform a pairplot of variables? </b>

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='time')
```

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.info()
```

```python
tips.head(3)
```

<b> How to perform a bar plot of variables? </b>

```python
sns.barplot(x='time',y='total_bill',data=tips) # total bill average
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator = np.std) # standard deviation
```

```python
sns.countplot(x='day',data=tips)
```

<b> How to perform a boxplot of variables? </b>

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

<b> How to perform a violin plot of variables? </b>

```python
sns.violinplot(x='day',y='tip',data=tips)
```

<b> How to perform a strip plot of variables? </b>

```python
sns.stripplot(x='day', y='total_bill',data=tips,jitter=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```
