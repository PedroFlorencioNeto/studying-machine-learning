<b> How to import? </b>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

<b> How to plot a graph? </b>

```python
x = np.linspace(0,5,11)
x
```

```python
y = x*x
```

```python
plt.plot(x,y)
```

<b> How to put parameters? </b>

```python
plt.plot(x,y,'r--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Title')
```

<b> How to plot multiple graphs? </b>

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

<b> How to plot with plt.figure? </b>

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.6,0.4,0.2])
axes1.plot(x,y)
axes2.plot(x,y)
```


<b> How to plot more than one graph in the same figure? </b>

```python
fig,ax = plt.subplots()
ax.plot(x,x**3,'r')
ax.plot(x,y,'g--')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
```

```python
fig,ax = plt.subplots(5,5)
```

```python
fig,ax = plt.subplots(5,5)
plt.tight_layout()
```

<b> How to save the graph? </b>

```python
fig.savefig('imagem.jpeg')
```

<b> How to do a scatter plot? </b>

```python
plt.scatter(x,y)
```
