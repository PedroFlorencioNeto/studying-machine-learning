<b> Solving Systems of Linear Equations

```python
# 4x + 0y + 0z = 96
# -3x + 8y + 3z = 192
# 0x + 0y + 12z = 96

# A.X = B

A = np.array([[4,0,0],[-3,8,3],[0,0,12]])
# X = [x,y,z]
B = np.array([96,192,96])

inv_A = np.linalg.inv(A) # Inverse Matrix
X = inv_A.dot(B)
X
```

<b> Eigenpairs of matrix

```python
autovalue,autovector = np.linalg.eig(A)
print(autovalue)
```

```python
print(autovector)
```
